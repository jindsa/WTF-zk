```python
def mod(a, b):
    remainder = a % b
    if remainder < 0:
        # 调整余数确保非负
        remainder += abs(b)
    return remainder
```

```python
a = 17
b = 5
remainder = mod(a, b)
print(f'{a} mod {b} = {remainder}')
# 17 mod 5 = 2

a = 25
b = 7
remainder = mod(a, b)
print(f'{a} mod {b} = {remainder}')
# 25 mod 7 = 4

a = 69
b = 23
remainder = mod(a, b)
print(f'{a} mod {b} = {remainder}')
# 69 mod 23 = 0
```

```python
# 二次剩余  枚举判断是否为2次剩余
# 判断ints中的数是否为n的二次剩余，并求出模n方程根

n = 29
ints = [14, 6, 11,28,27,26,25,24,23,22,21,20,19,18,17,16,15]
qr = [a for a in range(n) if pow(a, 2, n) in ints and pow(a, 2, n) != 0]
res = [pow(b,2,n) for b in qr]
print(list(qr))
print(list(res))


```
