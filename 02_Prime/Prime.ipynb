```python
# Sieve of Eratosthenes
def sieve_of_eratosthenes(n):
    primes = [True] * (n + 1)
    primes[0] = primes[1] = False
    for i in range(2, int(n**0.5) + 1):
        if primes[i]:
            for j in range(i * i, n + 1, i):
                primes[j] = False
    return [x for x in range(n + 1) if primes[x]]

# Example: Find all prime numbers less than or equal to 20
limit = 20
prime_numbers = sieve_of_eratosthenes(limit)
print(f'Prime numbers less than or equal to {limit}: {prime_numbers}')
# Prime numbers less than or equal to 20: [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
def mod_pow(base, exponent, modulus):
    result = 1
    base = base % modulus # 预处理 避免大数计算
    while exponent > 0: # 还有二进制位未处理
        if exponent % 2 == 1:  # 检查当前 最低位是否为 1 
            result = (result * base) % modulus
        base = (base * base) % modulus  # 底数平方 
        exponent //= 2  # 丢弃最低位  指数右移一位
    return result

mod_pow(7,4,13)
```
